```python
from hapi import *
import numpy as np
import matplotlib.pyplot as plt

# Initialize HITRAN database
db_begin('data')

# Gases and HITRAN molecule numbers
gases = {
    'H2O': 1,
    'CO2': 2,
    'O3': 3,
    'CO': 5,
    'CH4': 6,
    'NO2': 10,
}

# Wavenumber range (200–2000 nm = 50000 to 5000 cm⁻¹)
nu_min = 5000
nu_max = 50000

# Fetch HITRAN data for each gas, using the most abundant isotope (isotope 1)
for gas, ID in gases.items():
    fetch(gas, ID, 1, nu_min, nu_max)  # fixed: use isotope 1

# Wavenumber grid
step = 1.0
nu_grid = np.arange(nu_min, nu_max, step)

# Simulate absorption spectrum
nu, coeff = absorptionCoefficient_Voigt(
    SourceTables=list(gases.keys()),
    Diluent={'air': 1.0},
    Environment={
        'T': 296,
        'p': 1.0,
        'l': 1.0,
        'H2O': 0.02,
        'CO2': 0.0004,
        'O3': 2e-8,
        'CO': 1e-5,
        'CH4': 1.8e-6,
        'NO2': 1e-7,
    },
    OmegaGrid=nu_grid
)

# Convert wavenumber to wavelength (nm)
wavelength_nm = 1e7 / nu
wavelength_nm = wavelength_nm[::-1]
coeff = coeff[::-1]

# Plotting
plt.figure(figsize=(12, 5))
plt.plot(wavelength_nm, coeff)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorption Coefficient (cm⁻¹)')
plt.title('Simulated Absorption Spectrum (200–2000 nm) for Urban Gas Mixture')
plt.grid(True)
plt.xlim(200, 2000)
plt.tight_layout()
plt.show()
```
