```python
from hapi import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import Avogadro, Boltzmann, atmosphere

# Initialize database
db_begin('data')

# Parameters for NO2 (Nitrogen Dioxide)
molecule_name = 'NO2'       # Gas name for NO2
molecule_id = 6             # HITRAN ID for NO2
isotope_id = 1              # Use major isotope (NO2)
wavelength_min = 200        # nm
wavelength_max = 1500       # nm
resolution = 0.01           # cm^-1

# Convert nm to cm^-1
nu_min = 1e7 / wavelength_max  # cm^-1
nu_max = 1e7 / wavelength_min  # cm^-1

# Load lines from HITRAN
fetch(molecule_name, molecule_id, isotope_id, nu_min, nu_max)

# Gas and environmental parameters
pressure = 1.0     # atm
temperature = 296  # K
concentration = 100e-6  # mole fraction (e.g., 100 ppm for NO2)
path_length = 1.0  # cm

# Absorption coefficients
nu, coef = absorptionCoefficient_Voigt(
    SourceTables=molecule_name,
    Environment={'T': temperature, 'p': pressure},
    WavenumberStep=resolution,
    HITRAN_units=True
)

# Convert wavenumber to wavelength (nm)
wavelengths = 1e7 / nu

# Apply Beer-Lambert Law to get transmittance
number_density = (pressure * atmosphere) / (Boltzmann * temperature)  # molecules/m^3
number_density /= 1e6  # convert to molecules/cm^3
concentration_cm3 = number_density * concentration

absorbance = coef * concentration_cm3 * path_length
transmittance = np.exp(-absorbance)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(wavelengths, absorbance, label=f'{molecule_name} absorbance')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.title(f'Absorption Spectrum of {molecule_name} ({concentration*1e6:.0f} ppm)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
