```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  load parquet
spectra_data = pd.read_parquet("50k N2O spectra gen-1 noise.parquet")
spectra_data.head()
```

```python
#  visualise a sample spectrum
index = 1550
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(range(902), spectra_data["spectra_w_noise"].iloc[index])
ax1.set_title("with noise")
ax1.set_xticks(range(0, 1050, 100)) 
for x in range(0, 1050, 50):
    ax1.axvline(x=x, color='lightgray', linestyle='--', linewidth=0.5)

ax2.plot(range(902), spectra_data["spectra_no_noise"].iloc[index])
ax2.set_title("without noise")
plt.show()
```
